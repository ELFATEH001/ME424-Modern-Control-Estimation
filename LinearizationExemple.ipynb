```python
import numpy as np
import sympy as sym
```

```python
#define constants

t, g = sym.symbols('t g')
mc, mp = sym.symbols('m_c m_p')
L, fx = sym.symbols('L fx')

#define variables
theta, z = sym.symbols(r'\theta, z')
z_d = sym.symbols(r'\dot{z}')
theta_d = sym.symbols(r'\dot{\theta}')
```

```python

z_dd = 1/(mc + mp*sym.sin(theta)**2) * (fx + mp*sym.sin(theta)*(L*theta_d**2 + g*sym.cos(theta)))
theta_dd = 1/(L*(mc + mp*sym.sin(theta)**2)) * (-fx*sym.cos(theta) - mp*L*theta_d**2*sym.cos(theta)*sym.sin(theta) - (mc + mp)*g*sym.sin(theta))

x_vec = sym.Matrix([z, theta, z_d, theta_d]) 
vector_field = sym.Matrix([z_d, theta_d, z_dd, theta_dd])
```

```python
vector_field
```

```python
pfpx = vector_field.jacobian(x_vec)
pfpu = vector_field.jacobian([fx]) 
```

```python
pfpx
```

```python
# g = 9.81
# mc = 10.0
# mp = 1.0
# L = 0.5

pfpx_num = sym.lambdify((g, fx, mc, mp, L, theta, theta_d, z, z_d), pfpx)
pfpu_num = sym.lambdify((g, fx, mc, mp, L, theta, theta_d, z, z_d), pfpu)

Ahat = pfpx_num(9.81, 0.0, 10.0, 1.0, 0.5, np.pi, 0.0, 0.0, 0.0)
Bhat = pfpu_num(9.81, 0.0, 10.0, 1.0, 0.5, np.pi, 0.0, 0.0, 0.0)
```
