```python
import numpy as np
from pydrake.all import MathematicalProgram, Solve
```

```python
def minimize_quadratic_cost(Q, A, b):

    n = Q.shape[0]
    m = A.shape[0]

    prog = MathematicalProgram()
    
    p = prog.NewContinuousVariables(n, "p")

    prog.AddQuadraticCost(Q, np.zeros(n), p)

    prog.AddLinearEqualityConstraint(A, b, p)

    result = Solve(prog)

    if result.is_success():
        p_opt = result.GetSolution(p)
        optimal_value = result.get_optimal_cost()
        return p_opt, optimal_value
    else:
        print("Solver failed:", result.get_solver_details().info)
        return None, None
```

```python
T = 1
Q = np.array([[T**3/3, T**4/4],
                 [T**4/4, T**5/5]])
A = np.array([[1, 1]])
b = np.array([1])

minimize_quadratic_cost(Q, A, b)
```
