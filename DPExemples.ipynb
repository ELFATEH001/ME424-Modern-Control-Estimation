```python
costMatrix = np.array([[0, 4, -1, -1, -1,-1, 6,-1],
                       [4, 0, 3, -1, -1,-1,-1, 3],
                       [-1, 3, 0, 5, -1,-1,-1,-1],
                       [-1, -1, 5, 0, 2,-1,-1, 5],
                       [-1, -1, -1, 2, 0, 3,-1, 4],
                       [-1, -1, -1, -1, 3, 0, 4,-1],
                       [ 6, -1, -1, -1, -1, 4, 0, 2],
                       [-1, 3, -1, 5, -5, -1, 2, 0]])

costMatrix = costMatrix + np.eye(8) * 10

print(costMatrix)
```

```python
def value_iteration(Vfunc, costMatrix, xf):
    n = len(costMatrix)
    J = -1*np.ones(n)
    Vnext = -1*np.ones(n)
    mustar = -1*np.ones(n)

    for xi in range(n):
        if xi == xf:
            Vnext[xi] = Vfunc[xi]
            mustar[xi] = xi
            continue
        for ui in range(n):
            if costMatrix[xi,ui] != -1:
                J[ui] = costMatrix[xi,ui] + Vfunc[ui]
            else:
                J[ui] = 10**6
        Vnext[xi] = np.min(J)
        mustar[xi] = np.argmin(J)
    return Vnext, mustar
    
```

```python
n  = 8 
N  = 4
xf = 3
g  = 10**6*np.ones(n)
g[xf] = 0

V  = np.zeros((N,n))
mu = np.zeros((N,n))
V[0,:] = g

for j in range(N-1):
    V[j+1,:], mu[j+1,:] = value_iteration(V[j,:], costMatrix, xf)
```

```python
print("V = ", V)
print("mu = ", mu)
```
