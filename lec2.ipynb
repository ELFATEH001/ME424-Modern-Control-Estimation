```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

A = np.array([[-0.2, 1], [0, -1]])
B = np.array([[-1], [1]])
C = np.array([[1, 0]])
D = np.array([[0]])

sys = signal.StateSpace(A, B, C, D)
print(sys)
```

```python
t = np.linspace(0, 50, 100)
x0 = [-1, 1]
t,y = signal.step(sys, T=t, X0=x0)

plt.plot(t, y)
```

```python
# now let's take a look at disctization
dt = 0.1
sysd = sys.to_discrete(dt, method='euler')
print(sysd)
```

```python
#simulate the discrete system
t, y = signal.dstep(sysd, x0=x0, n=500)
plt.plot(t, np.squeeze(y))
```
